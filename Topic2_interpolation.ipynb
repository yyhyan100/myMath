### Lagrange Interplation

```python
import numpy as np
from myMath.interpolation import intp_1d
intp=intp_1d()

x=np.array([1.0,2,3,5,6,8, 12.0])
y=np.array([0.2,0.4,0.7,1.1,2.1,4.5,9.5])
x_input=4.5

print(intp.Lagrange(x,y,x_input))

# for multple points interpolation, use numpy array!
x_input=np.array([1.2,4.5,7.3])
print(intp.Lagrange(x,y,x_input))
```

### Newton's Divided Difference

```python
import numpy as np
from myMath.interpolation import intp_1d
intp=intp_1d()

x=np.array([1.0,2,3,5,6,8, 12.0])
y=np.array([0.2,0.4,0.7,1.1,2.1,4.5,9.5])
x_input=4.5

print(intp.NewtonDD(x,y,x_input))

# for multple points interpolation, use numpy array!
x_input=np.array([1.2,4.5,7.3])
print(intp.NewtonDD(x,y,x_input))
```

### Hermite Interpolation

```python
# 2 pnts formula
import numpy as np
from myMath.interpolation import intp_1d
intp=intp_1d()
x=np.array([30,45,60])
y=np.array([0.5,0.707107,0.866025])
yp=np.array([0.866025,0.707107,0.5])
x_input=40
print(intp.Hermite_2pts(x[:2],y[:2],yp[:2],x_input))
```

```python
# 3 pnts formula
import numpy as np
from myMath.interpolation import intp_1d
intp=intp_1d()
x=np.array([30,45,60])
y=np.array([0.5,0.707107,0.866025])
yp=np.array([0.866025,0.707107,0.5])
x_input=40
print(intp.Hermite_3pts(x,y,yp[1],x_input))
```

### spline interpolation

```python
import numpy as np
from myMath.interpolation import intp_1d
intp=intp_1d()

x=np.array([1.0,2,3,5,6,8, 12.0])
y=np.array([0.2,0.4,0.7,1.1,2.1,4.5,9.5])
x_input=4.5

print(intp.spline(x,y,x_input,0))

# for multple points interpolation, use numpy array!
x_input=np.array([1.2,4.5,7.3])
print(intp.spline(x,y,x_input))
```
